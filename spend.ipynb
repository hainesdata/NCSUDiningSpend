```python
import pandas as pd

raw = pd.read_csv('data/Fall Commuter $800_statement_2021-04-01_to_2021-10-30.csv')
wrk = raw.copy()
wrk.Date = pd.to_datetime(wrk.Date)
wrk
```

```python
import plotly.express as px

fig = px.scatter(wrk, x='Date', y='Balance', trendline='lowess')
fig.show()
```

```python
wrk = wrk[~(wrk.Description.str.contains('Tax') | wrk.Description.str.contains('Credit'))]
descriptions = wrk.Description.value_counts().reset_index()
fig2 = px.bar(descriptions, x='index', y='Description')
fig2.show()
```

```python
spend = wrk.groupby(["Description"]).Amount.sum().abs().reset_index()

fig3 = px.pie(spend, values='Amount', names='Description')
fig3.show()
```
